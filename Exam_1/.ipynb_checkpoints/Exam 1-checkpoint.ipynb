```python
VarNames = ["EventId","DER_mass_MMC","DER_mass_transverse_met_lep","DER_mass_vis","DER_pt_h","DER_deltaeta_jet_jet","DER_mass_jet_jet","DER_prodeta_jet_jet","DER_deltar_tau_lep","DER_pt_tot","DER_sum_pt","DER_pt_ratio_lep_tau","DER_met_phi_centrality","DER_lep_eta_centrality","PRI_tau_pt","PRI_tau_eta","PRI_tau_phi","PRI_lep_pt","PRI_lep_eta","PRI_lep_phi","PRI_met","PRI_met_phi","PRI_met_sumet","PRI_jet_num","PRI_jet_leading_pt","PRI_jet_leading_eta","PRI_jet_leading_phi","PRI_jet_subleading_pt","PRI_jet_subleading_eta","PRI_jet_subleading_phi","PRI_jet_all_pt","Weight","Label"]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename="/home/zachary.hinkle/DLHEP2018/Exam_1/mytraining.csv"
df = pd.read_csv(filename, names=VarNames)
```

```python
#This cell removes all rows that contain a value of -999;
#however, I didn't have time to change my code to handle
# computing the AMS in this case. As a cursory metric,
#the auc's were worse for all the classifiers

#for col in df:
#    df = df[df != -999]
#df = df.dropna()
```

```python
for col in df:
    if np.any(df[col] == -999):
        df = df.drop(col, axis=1)
```

```python
df.loc[df.Label == 's', 'Label'] = 1
df.loc[df.Label == 'b', 'Label'] = 0
df.Label = df.Label.astype('int')
```

```python
df
```

```python
GoodVars = []
for col in df:
    if col in VarNames:
        GoodVars.append(col)
print GoodVars
```

```python
df_sig=df[df.Label==1]
df_bkg=df[df.Label==0]
```

```python
for var in GoodVars:
    print col
    plt.figure()
    plt.hist(df_sig[col],bins=100,histtype="step", color="red",label="background",stacked=True)
    plt.hist(df_bkg[col],bins=100,histtype="step", color="blue", label="signal",stacked=True)
    plt.legend(loc='upper right')
    plt.show()
```

```python
N_Train=20000

Train_Sample=df[:N_Train]
Test_Sample=df[N_Train:]

X_Train=Train_Sample[GoodVars[1:-2]]
y_Train=Train_Sample["Label"]

X_Test=Test_Sample[GoodVars[1:-2]]
y_Test=Test_Sample["Label"]

Test_sig=Test_Sample[Test_Sample.Label==1]
Test_bkg=Test_Sample[Test_Sample.Label==0]
```

```python
import sklearn.discriminant_analysis as DA
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn import linear_model
LDA=DA.LinearDiscriminantAnalysis()
PA = PassiveAggressiveClassifier(random_state=0)
SGD = linear_model.SGDClassifier()
```

```python
LDA.fit(X_Train,y_Train)
PA.fit(X_Train,y_Train)
SGD.fit(X_Train,y_Train)
```

```python
#This cell defines a function that takes a classifier as input and then
# plots the histograms of its classified signal and bkg for the test sig and bkg
#It then makes a second graph of the roc curve
def MakeGraph(clsf):
    from sklearn.metrics import roc_curve, auc
    plt.figure(figsize=(20,4))

    #histogram of classified signal and bkg
    plt.subplot(1,3,1)
    plt.hist(clsf.decision_function(Test_sig[GoodVars[1:-2]]),bins=100,histtype="step", color="blue", label="signal",stacked=True)
    plt.hist(clsf.decision_function(Test_bkg[GoodVars[1:-2]]),bins=100,histtype="step", color="red", label="background",stacked=True)
    plt.legend(loc='upper right')

    plt.show()
    
    ###ROC Curve Plot####
    plt.figure()
    fpr, tpr, _ = roc_curve(y_Test, clsf.decision_function(X_Test))
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr,tpr,color='blue',label='ROC curve for raw+features (area = %0.2f)' % roc_auc)

    plt.legend(loc="lower right")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

    plt.show()
    
```

```python
MakeGraph(LDA)
MakeGraph(PA)
MakeGraph(SGD)
```

```python
def MakeAMS(pred,true):
    s = 0; sBest = 0;
    b = 0; bBest = 0;
    br = 10;
    for i in range(0,len(pred)):
        if pred[i] == 1:
            if true[i+N_Train] == 1:
                s += df.Weight[i+N_Train]
            else:
                b += df.Weight[i+N_Train]
                
        if true[i+N_Train] == 1:
            sBest += df.Weight[i+N_Train]
            
    AMS = np.sqrt(2*((s+b+br)*np.log(1+s/(b+br))-s))
    AMSBest = np.sqrt(2*((sBest+bBest+br)*np.log(1+sBest/(bBest+br))-sBest))
    return [AMS, AMSBest]
    
```

```python
print MakeAMS(LDA.predict(X_Test),y_Test)
print MakeAMS(PA.predict(X_Test),y_Test)
print MakeAMS(SGD.predict(X_Test),y_Test)
```

```python
from sklearn.externals import joblib
joblib.dump(LDA, 'LDA.pkl')
joblib.dump(PA, 'PA.pkl')
joblib.dump(SGD, 'SGD.pkl')
```
